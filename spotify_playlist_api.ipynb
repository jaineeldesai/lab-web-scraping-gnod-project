## Songs from Spotify Playlist with Audio Features

### For clustering our songs we will need:

- title of the song
- artist(s) of the song
- the song's audio features

### Connecting to the spotify API and authentication with credentials

```python
import pandas as pd 
from pandas import json_normalize
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
pd.set_option('display.max_columns', None)
```

```python
secrets_file=open('secrets.txt','r')
```

```python
string=secrets_file.read()
```

```python
secrets_dict={}
for line in string.split('\n'):
    if len(line) > 0:
        secrets_dict[line.split(':')[0]]=line.split(':')[1].strip()
```

```python
#Initialize SpotiPy with user credentials
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=secrets_dict['clientid'],
                                                           client_secret=secrets_dict['clientsecret']))
```

```python
from random import randint
from time import sleep

def get_playlist_tracks(playlist_id):
    results = sp.user_playlist_tracks("spotify",playlist_id)
    tracks = results['items']
    while results['next']!=None:
        results = sp.next(results)
        tracks = tracks + results['items']
        sleep(randint(1,3000)/1000) # respectful nap
    return tracks
```

```python
all_tracks=get_playlist_tracks('1G8IpkZKobrIlXcVPoSIuf')
```

```python
len(all_tracks)
```

```python
tracks=json_normalize(all_tracks)
```

```python
tracks.head()
```

```python
tracks['track.name'][0] #title seems okay
```

```python
tracks['track.artists'][0] 

#it is a list of dict; instead we will make dict of the list as it is easier to handle
```

```python
tracks['track.artists'].isna().sum()
```

```python
tracks.isna().sum()
```

```python
#dropping data where we dont have track names i.e. track.name= NaN

tracks=tracks[tracks['track.name'].isna()==False]
```

```python
len(tracks)
```

```python
def list_to_dict(x):
    return {i: x[i] for i in range(len(x))}
            
tracks['artist_dict'] = tracks['track.artists'].apply(list_to_dict)
tracks.head()
```

```python
# we can even create a new column containing these artist_dicts as dataframes

def expand_list_dict2(row):
    df = pd.DataFrame.from_dict(row['artist_dict'], orient='index')
    df['song_id'] = row['track.id']
    return df

tracks['artists_dfs'] = tracks.apply(expand_list_dict2, axis=1)
tracks['artists_dfs'][0]
```

```python
# now we create a new dataframe with all these artists

artist_df = pd.DataFrame(columns=['external_urls', 'href', 'id', 'name', 'type', 'uri'])
for mini_df in tracks['artists_dfs']:
    artist_df = pd.concat([artist_df, mini_df], axis=0)
    
artist_df.head()
```

```python
#merging the artist_df with tracks

df_merged= pd.merge(left=tracks,
                    right=artist_df,
                    how='inner',
                    left_on='track.id',
                    right_on='song_id')
df_merged.head()
```

```python
df_merged.isna().sum()
```

```python
len(df_merged)
```

```python
#dropping songs that dont have song_id

df_merged=df_merged[df_merged['song_id'].isna()==False]
```

```python
df_final = df_merged[['track.name', 'name', 'song_id']]
```

```python
#need to reset_index as we dropped some data
df_final.reset_index(drop=True,inplace=True)
```

```python
df_final
```

### Retrieving Audio Features

```python
#we will use song_id to retrieve the audio features of a song

chunks = [(i, i+100) for i in range(0, len(df_final), 100)]
audio_features_list=[]
for chunk in chunks:
    id_list100 = df_final['song_id'][chunk[0]:chunk[1]]
    audio_features_list = audio_features_list + sp.audio_features(id_list100)
    sleep(randint(1,3000)/1000)
len(audio_features_list)
```

```python
audio_features_df = json_normalize(audio_features_list)
```

```python
#need to drop duplicates as we have same id as song_id for each artist

audio_features_df.drop_duplicates(inplace=True)
audio_features_df
```

```python
#merging the track dataframe with audiofeatures

df_w_audio_ft = pd.merge(left=df_final,
                        right=audio_features_df,
                        how='inner',
                        left_on='song_id',
                        right_on='id')
df_w_audio_ft
```
